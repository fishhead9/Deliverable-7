```python
n = int(input("Enter an an integer value greater than 0: "))
```

```python
E = even = []
O = odd = []
F = fibo = [0,1]

for i in range(n):
  even.append(2*i)
  odd.append(2*i+1)
  if i > 1: fibo.append(fibo[i-2]+fibo[i-1])
```
